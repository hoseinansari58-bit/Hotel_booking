# Extract — Load the raw CSV

Section 1 content

```python


import pandas as pd
import os

#extracting raw data
file_path = os.path.join("..", "Data", "raw_data", "hotel_booking.csv")

#loading raw data into a dataframe
df_raw = pd.read_csv(file_path)
df_raw.head()
```

---

# Transform — Clean and reshape the data

```python


#Droping PII columns from the dataset (GDPR Compliance)
df_clean = df_raw.drop(columns=['name', 'email', 'phone-number', 'credit_card'])

#Removing Duplicates

df_clean = df_clean.drop_duplicates()


#Filling missing values

df_clean = df_clean.fillna({
    'children': 0,
    'country': 'Unknown',
    'distribution_channel': 'Unknown',
    'agent': 0,
    'company': 'Unknown',
    'customer_type': 'Unknown',
})


df_clean.index = df_clean.index.astype(str)

#Converting date column to datetime
df_clean['reservation_status_date'] = pd.to_datetime(df_clean['reservation_status_date'])

#Cleaning column names
df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)

# Rebuild the Index object to avoid pandas formatting bugs
df_clean.columns = pd.Index(df_clean.columns)

#Resetting index to avoid the pandas to_csv bug

df_clean = df_clean.reset_index(drop=True)

#Saving cleaned data to clean_folder

clean_path = os.path.join("..", "Data", "clean_folder", "hotel_booking_clean.csv")
os.makedirs(os.path.join("..", "Data", "clean_folder"), exist_ok=True)
df_clean.to_csv(clean_path, index=False)


print("Cleaned dataset saved successfully!")

```
