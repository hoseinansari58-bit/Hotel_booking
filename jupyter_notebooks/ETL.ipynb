# Extract — Load the raw CSV

Section 1 content

```python

import pandas as pd
import os

file_path = os.path.join("Data", "raw_data", "hotel_booking.csv")

df_raw = pd.read_csv(file_path)
df_raw.head()
```

---

# Transform — Clean and reshape the data

Section 2 content

```python
#droping columns that holds personal detals- 'name', 'email', 'phone-number', 'credit_card' (for privacy reasons- GDPR compliance)
df_clean = df_raw.drop(columns=['name', 'email', 'phone-number', 'credit_card'])

#removing the duplicates
df_clean =df_clean.drop_duplicates()

df_clean = df_clean.fillna({
     'children': 0,
     'country': 'Unknown',
     'distribution_channel': 'Unknown',
     'agent': 0,
     'company': 'Unknown',
     'customer_type': 'Unknown',
})    

df_clean.index.astype(str)
df_clean.head()
```

---

```python
df_clean.columns = df_clean.columns.str.strip().str.lower().str.replace(' ', '_')
df_clean.head()
```

#### Converting the date column to datetime format and checking for any missing values

```python
df_clean['reservation_status_date'] = pd.to_datetime(df_clean['reservation_status_date'])
df_clean.isnull().sum()
```
